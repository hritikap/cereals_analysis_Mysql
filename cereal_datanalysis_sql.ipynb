### What is the average content of each nutrient and products(cereals) per manufacturer?

```python
%sql select mfr, avg(calories) as average_cal, avg(protein) as average_protein,avg(fat) as average_fat,avg(sodium) as average_sodium,avg(fiber) as average_fiber,avg(carbo) as average_carbo,avg(sugars) as average_sugars,avg(potass) as average_potass,avg(vitamins) as average_vitamins from cereals group by mfr
```

### Energy content of each cereal according to protein, fat, and carbohydrate data.

Research has shown how much energy our body can make from these macronutrients. We know that:

> - 1 gram of fats provides 9 kcal of energy
> - 1 gram of carbohydrates = 4 kcal
> - 1 gram of fibers = 2 kcal
> - 1 gram of polyols = 2,4 kcal
> - 1 gram protein = 4 kcal
> - 1 gram of alcohol = 7 kcal


Using the conversion factors above, we can calculate the energy content(kcal/100g) of your food by multiplying these values by the amount of each of the macronutrients present.

```python
%%sql
ALTER TABLE cereals
ADD COLUMN energy_content INT;
```

```python
%%sql
UPDATE cereals
SET energy_content=(fat*9 +carbo*4 + protein*4)
```

```python
%%sql
ALTER TABLE cereals
DROP energy;
```

### Which manufacturer possesses the best shelf location?

```python
%%sql
select mfr,count(shelf) from cereals where shelf=2 group by mfr
```

Research has shown that most consumers believe the middle shelf - or the shelf closest to their eye level - is the best placement for products in a store resulting as the best shelf location.According to our dataset '2' is considered as the middle self location so we group the manufacturer by counting the manufacturer products who have the shelf location of '2'.

So,from the above analysis it is found that the manufacturer kellogs(K) and General Mills(G) has the highest number of middle shelf count so it posseses the best shelf location
